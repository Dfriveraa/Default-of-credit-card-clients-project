```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
import joblib
```

```python
df=pd.read_csv('./Database/DB_v1.csv',index_col=False)
df=df.drop(['Unnamed: 0'],axis=1)
y=df['DEF_PAY']
X=df.drop(['DEF_PAY'],axis=1)
```

```python
df
```

```python
scaler=StandardScaler()

svc = SVC(C=0.1, class_weight='balanced',probability=True, gamma=0.01)
mlp = MLPClassifier(activation='tanh', alpha=0.01, early_stopping=True,
              hidden_layer_sizes=70, max_iter=100)
rf=RandomForestClassifier(class_weight='balanced', criterion='entropy',
                       max_features=4, n_estimators=250,n_jobs=-1)

clf = VotingClassifier(estimators=[('svc', svc), ('rf', mlp), ('gnb', mlp)], voting='hard',n_jobs=-1)

pipeline = Pipeline([
      ('scaler', scaler),
      ('classification', clf)
    ])
features=[0, 1, 4, 5, 6, 7, 8, 10, 11, 12, 13, 15, 17, 18, 19, 20]
X=X.iloc[:,features]
pipeline.fit(X, y)

# Export the classifier to a file
joblib.dump(pipeline, 'PipelineFinal.joblib')
```

```python
aux=pipeline[1].transform(pipeline[0].transform(X))
```
