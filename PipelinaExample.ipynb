```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.neural_network import MLPClassifier
import joblib
```

```python
df=pd.read_csv('./Database/DB_v1.csv',index_col=False)
df=df.drop(['Unnamed: 0'],axis=1)
y=df['DEF_PAY']
X=df.drop(['DEF_PAY'],axis=1)
```

```python
scaler=StandardScaler()
clf=MLPClassifier(activation='tanh', alpha=0.1, early_stopping=True,
              hidden_layer_sizes=100, max_iter=300)

pipeline = Pipeline([
      ('scaler', scaler),
      ('classification', clf)
    ])

pipeline.fit(X, y)

# Export the classifier to a file
joblib.dump(pipeline, 'modelExample.joblib')
```

```python
pipe=joblib.load('modelExample.joblib')
```

```python
type(list(pipe.predict(X)))
```
